### Missing values

```python
# Finding out columns with missing values
df.isnull().sum()
```

```python
column_name = 'birth_year'
```

```python
# Calculate the mode of the column
mode_value = df[column_name].mode()[0]
```

```python
# Replace missing values with the mode
df[column_name] = df[column_name].fillna(mode_value)
```

```python
df.isnull().sum()
```

```python
#changing the values in gender column from 0 to not available.
df['gender'] = df['gender'].replace({0: 'Not Available'})
```
